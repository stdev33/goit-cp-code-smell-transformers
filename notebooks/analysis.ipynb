## Structure of the SmellyCode++ Dataset

Dataset contains 107k Java classes annotated for 4 types of code smells:
- Long Method
- God Class
- Feature Envy
- Data Class

Every record contains the code text (`Code`) and corresponding binary labels indicating the presence or absence of each code smell type.
